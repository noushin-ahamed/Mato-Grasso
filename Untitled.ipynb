```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('NDVI_TRMM.csv')
```

```python
df.corr().round(3)
```

```python
sns.regplot(x="NDVI",
           y="TRMM",
           data=df)

plt.title(('Yearly NDVI vs TRMM Mato Grasso (2000-2020)'))
plt.show()
```

```python
lat_mod = smf.ols("TRMM ~ NDVI", data=df).fit()
print (lat_mod.summary())
```

```python
plt.hist(lat_mod.resid_pearson, 
         edgecolor='black',
         bins=[-2,-1.5,-1,-0.5,0,0.5,1,1.5,2])
plt.xlabel('Residuals')
```

```python
plt.plot(lat_mod.fittedvalues, lat_mod.resid_pearson,
         'oC0',  
         markeredgecolor='black')
plt.axhline(c = 'red')
plt.xlabel('Predicted y Values')
plt.ylabel('Standardized Residuals (z)')  
```

```python
qqfig = sm.qqplot(lat_mod.resid_pearson, line='r')
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(lat_mod, fig=fig)
```
